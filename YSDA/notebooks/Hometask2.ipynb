### Function for visualization

```python
def visualize_classification(classifier, X, y):
    from sklearn.base import clone
    figure(figsize=[10, 8])
    # clone classifier and train it
    classifier = clone(classifier)
    classifier.fit(X, y)
    # we will totally predict 200 * 200 = 40000 points
    x_points = numpy.linspace(X[:, 0].min() - 1, X[:, 0].max() + 1, 200)
    y_points = numpy.linspace(X[:, 1].min() - 1, X[:, 1].max() + 1, 200)
    grid_x, grid_y = numpy.meshgrid(x_points, y_points)
    grid = numpy.c_[grid_x.ravel(), grid_y.ravel()]
    # Let's evaluate it!
    grid_predictions = classifier.predict_proba(grid)
    signal_probability = grid_predictions[:, 1]
    # Reshape & mirror
    XX = grid_predictions.reshape([200, 200, 3])[::-1, :, :]
    imshow(XX, extent=[x_points[0], x_points[-1], y_points[0], y_points[-1]])
    scatter(X[:, 0], X[:, 1], c=numpy.take(['r', 'g', 'b'], y), s=50, cmap='spring')
    xlim(x_points[0], x_points[-1])
    ylim(y_points[0], y_points[-1])
    show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
X, y = make_blobs(n_samples=300, n_features=2, centers=[[0, 0], [1, 2], [-1, -2]])
visualize_classification(KNeighborsClassifier(), X, y)
```

Please don't forget to train/test split

```python
from sklearn.metrics import roc_curve
def roc_blue_vs_all(classifier, X_test, y_test, selected_class=0):
    """
    classifier should be already trained
    Zero corresponds to first (blue) class,    
    """
    probability_blue = classifier.predict_proba(X_test)[:, selected_class]
    fpr, tpr, thresholds = roc_curve(y_test == selected_class, probability_blue)
    plot(fpr, tpr)
    xlabel('non-blue efficiency')
    ylabel('blue efficiency')
```
