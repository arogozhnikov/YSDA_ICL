### Let's introduce the function that will plot how particular classifier works

You can use it with any 2-dimensional dataset and nearly any classifier

```python
def visualize_classification(classifier, X, y):
    from sklearn.base import clone
    figure(figsize=[10, 8])
    # clone classifier and train it
    classifier = clone(classifier)
    classifier.fit(X, y)
    # we will totally predict 200 * 200 = 40000 points
    x_points = numpy.linspace(X[:, 0].min() - 1, X[:, 0].max() + 1, 200)
    y_points = numpy.linspace(X[:, 1].min() - 1, X[:, 1].max() + 1, 200)
    grid_x, grid_y = numpy.meshgrid(x_points, y_points)
    grid = numpy.c_[grid_x.ravel(), grid_y.ravel()]
    # Let's evaluate it!
    grid_predictions = classifier.predict_proba(grid)
    signal_probability = grid_predictions[:, 1]
    pcolormesh(x_points, y_points, signal_probability.reshape([200, 200]), cmap='spring')
    scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='spring')
    xlim(x_points[0], x_points[-1])
    ylim(y_points[0], y_points[-1])
    show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
for k in [1, 2, 3, 5, 10]:
    visualize_classification(KNeighborsClassifier(n_neighbors=k), X, y)
```

## Minkowski distance with different $p$

```python
from sklearn.neighbors import DistanceMetric
```

```python
for metric in ['euclidean', 'manhattan', 'canberra']:
    visualize_classification(KNeighborsClassifier(n_neighbors=5, metric=metric), X, y)
```

## Working with scaled features 
pay attention at scale of x-axis

```python
for factor in [0.1, 1., 10.]:
    X_new = X.copy()
    # multiply x-axis by factor
    X_new[:, 0] *= factor
    visualize_classification(KNeighborsClassifier(n_neighbors=5), X_new, y)
```

# What is the optimal k in kNN

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)
```

## Estimating the accuracy (part of correctly classified events)

```python
from sklearn.metrics import accuracy_score
scores = []
for k in range(1, 51):
    y_pred = KNeighborsClassifier(n_neighbors=k).fit(X_train, y_train).predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    scores.append(accuracy)
plot(scores)
```

this doesn't really mean we have information about which k to use. <br />

Let's use default `grid_search` to check this by cross-validation

```python
from sklearn.grid_search import GridSearchCV
params = {'n_neighbors': range(1, 51)}
grid_searcher = GridSearchCV(KNeighborsClassifier(), params, cv=5)
grid_searcher.fit(X, y)
```

# Visualizing (mean - std, mean + std)

```python
means = []
stds = []
for elem in grid_searcher.grid_scores_:
    means.append(np.mean(elem.cv_validation_scores))
    stds.append(sqrt(np.var(elem.cv_validation_scores)))
means = np.array(means)
stds = np.array(stds)
```

```python
figure(figsize=(8, 6))
plot(means)
fill_between(range(len(means)), means + stds, means - stds, alpha = 0.1, facecolor='blue')
xlabel('Number of nearest neighbours')
ylabel('Accuracy')
```
